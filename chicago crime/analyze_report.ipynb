```python
# Step 1: Fill NaN values by Grouping by 'Community Area' and filling with Mode
def fill_with_mode(group):
    # Check if mode() returns a valid value
    mode_value = group.mode()
    if not mode_value.empty:
        return group.fillna(mode_value[0])  
    else:
        return group  

# Apply the function to fill NaN values within each group
cri['Ward'] = cri.groupby('Community Area')['Ward'].transform(fill_with_mode)

# Step 2: For any remaining NaN values, fill them with the Mode of the entire dataset
cri['Ward'].fillna(cri['Ward'].mode()[0], inplace=True)


```

```python
# Create a mapping of Description to most common Location Description
location_mapping = cri.groupby('Description')['Location Description'].agg(
    lambda x: x.value_counts().index[0] if len(x.dropna()) > 0 else None
).to_dict()

# Remove None values from mapping
location_mapping = {k: v for k, v in location_mapping.items() if v is not None}

# Fill NaNs using the mapping
cri['Location Description'] = cri.apply(
    lambda row: location_mapping.get(row['Description'], row['Location Description'])
    if pd.isna(row['Location Description'])
    else row['Location Description'],
    axis=1
)

# Fill any remaining NaNs with overall mode
overall_mode_location = cri['Location Description'].dropna().mode()[0] if not cri['Location Description'].dropna().empty else "Unknown"
cri['Location Description'].fillna(overall_mode_location, inplace=True)
```
